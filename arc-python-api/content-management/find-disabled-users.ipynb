# Find Disabled Users

Find all disabled users in the org and count the number of items owned by each disabled user. Then, print a list of items and users.

```python
import pandas as pd
from IPython.display import display

import arcgis
from arcgis.gis import GIS
```

#### Connect to ArcGIS Online

```python
c3GIS = GIS(profile='aolson_prfl2')
c3GIS.users.me
```

## Get list of all users

```python
all_users = c3GIS.users.search(query='*',max_users=666)
print(f'Org has {len(all_users)} users')
```

## Create subset list of disabled users

```python
disabled_users = [user for user in all_users if user.disabled == True]
print(f'Org has {len(disabled_users)} disabled users')
disabled_users
```

## Count items owned by disabled users

```python
# Return content for each disabled user in the org, compile into single snapshot
disabled_content = []

for user in disabled_users:
    query = f'owner: {user.username}'
    user_content = c3GIS.content.advanced_search(query=query, max_items=666)['results']
    disabled_content += user_content

print(f'{len(disabled_content)} items found for disabled users')
```

## Analyze disabled user's content with Pandas

```python
# Create dataframe
content_df = pd.DataFrame(disabled_content)
print(f'Fieldnames: {list(content_df.columns)}')
content_df.head()
```

```python
# Use value_counts() to see how many items you have with a particular key:value pair
# content_df.type.value_counts().head(10) # Return top 10 most frequent item types
content_df.type.value_counts()
```

```python
# Summary count of owner item content
content_df.owner.value_counts()
```

```python
# Summary count of access level for items
content_df.access.value_counts()
```

```python
# Group item types by user
# content_df.groupby('owner').type.value_counts().head(10)
content_df.groupby('owner').type.value_counts()
```

```python
# Group items by type and their access level
content_df.groupby('type').access.value_counts()
```
